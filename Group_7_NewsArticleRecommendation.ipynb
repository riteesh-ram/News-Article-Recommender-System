```python
 
#Importing the required libraries
import requests
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pickle
 
#Function to get html of a webpage
def getWebsite(url):
  req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
  webpage = urlopen(req).read()
  soup = BeautifulSoup(webpage, "html.parser")
  return soup
 
#Lists to store the data
headlines = []
content = []
date_of_creation=[]
 
#Initial webpage
links = ["https://www.hindustantimes.com/latest-news"]
 
#Main code
for _ in links:
  soup = getWebsite(_)
  articles = soup.find_all('h2', class_ = 'hdg3')
 
  for article in articles:
    try:
      headlines.append(article.get_text())
      subsoup = getWebsite("https://www.hindustantimes.com" + article.a.get('href'))
      if subsoup.find('div', class_ = 'detail') != None:
        content.append(subsoup.find('div', class_ = 'detail').get_text())
        date_of_creation.append(subsoup.find("div",class_="dateTime").get_text())
    except:
      continue
 
  if len(links) < 40 and soup.find('li', class_ = 'next').a != None:
    links.append("https://www.hindustantimes.com" + soup.find('li', class_ = 'next').a.get('href'))
 
print("Total Headlines: " + str(len(headlines)))
print("Total Articles: " + str(len(content)))
 
#Code to store the data extracted. 
with open("/content/drive/MyDrive/NLP_Recommender/headlinesFile.data", "wb") as filehandle:
  pickle.dump(headlines, filehandle)
 
with open("/content/drive/MyDrive/NLP_Recommender/corpusFile.data", "wb") as filehandle:
  pickle.dump(content, filehandle)
 
with open("/content/drive/MyDrive/NLP_Recommender/date_of_creationFile.data", "wb") as filehandle:
  pickle.dump(date_of_creation, filehandle)
```

```python
def preprocess_text(text):
  sen=re.sub('["/n","/t"]',"",text) #removing newlines
  sen=re.sub('[^0-9a-zA-Z]', ' ', text) #removing numbers and puncutations
  sen=re.sub(r"\s+[a-zA-Z]\s+", ' ',text) #removing single characters
  sen=re.sub(r"\s+"," ",text) #removing multiple spaces
  return sen
```

```python
from nltk import PorterStemmer
stemmer = PorterStemmer()

def lemmatize_stemming(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))

def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
```

```python
import random
random.seed(3)

def dummy_data(userNo,topicNo):
  interested_topics_num = []
  mu = topicNo/4
  var = topicNo/7
  for i in range(userNo): 
    temp = int(abs(random.gauss(mu,var)+1))          #number of topics interested with mu = topicNo/3 and variance = topicNo/7
    interested_topics_num.append(temp)
  user_interested_topic = []
  for i in range(userNo): 
    temp = random.sample(range(0, topicNo-1), interested_topics_num[i])          #'interested_topics_num' number of unique topics are generated for each user
    user_interested_topic.append(temp)
  user_topic_matrix=[]
  for i in range(userNo):
    sum = 0
    temp = []
    for j in range(topicNo):
      if j in user_interested_topic[i]:
        temp_val = random.randint(1, 100)
        sum +=temp_val
        temp.append(temp_val)
      else:
        temp.append(0)
    tempo = []
    for i in temp:
      if i == 0:
        tempo.append(0)
      else:
        tempo.append(i/sum)
    user_topic_matrix.append(tempo)
  return user_topic_matrix

user_topic_matrix = dummy_data(1000,20)
#user_topic_matrix
```

```python
def doc_topic_assign(content_data,topicNo):
  bow_vector = dictionary.doc2bow(preprocess(content_data))
  topicss = []
  indexes = []
  ret = []
  for index, score in lda_model_tfidf[bow_vector]:
    indexes.append(index)                #, key=lambda tup: -1*tup[1]:
    topicss.append(score)
  for i in range(topicNo):
    if i in indexes:
      k = indexes.index(i)
      ret.append(topicss[k])
    else:
      ret.append(0)  
  return ret


doc_topic_matrix = []
Topic_NUMB = 20
for i in range(data_text['merged'].size):
  doc_topic_matrix.append(doc_topic_assign(data_text['merged'][i],Topic_NUMB))

doc_topic_matrix
```

```python
import operator
def user_docs_sim(user, doc_topic_matrix):
  docs_size = doc_topic_matrix.shape[0]
  sim = {}
  for i in range(0, docs_size):
    sim[i] = float(cosine_similarity(user.reshape(1,-1), doc_topic_matrix[i].reshape(1,-1)))
  sim = dict( sorted(sim.items(), key=operator.itemgetter(1),reverse=True))
  return sim
```
